# 2. Transfromasi Dasar dengan DataFrames

Tugas 2

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# Membuat SparkSession
spark = SparkSession.builder.appName('HandsOnPertemuan3').getOrCreate()

# Data awal + tambahan
data = [
    ('James', 'Sales', 3000),
    ('Michael', 'Sales', 4600),
    ('Robert', 'Sales', 4100),
    ('Maria', 'Finance', 3000),
    ('Rani', 'Finance', 3300),
    ('Ali', 'IT', 4000),
    ('Siti', 'IT', 4500),
    ('Budi', 'HR', 3900),
    ('Citra', 'HR', 3500),
    ('Dina', 'Marketing', 4200),
    ('Andi', 'Marketing', 3700)
]

# Kolom DataFrame
columns = ['EmployeeName', 'Department', 'Salary']

# Membuat DataFrame
df = spark.createDataFrame(data, schema=columns)

# Menampilkan isi DataFrame
df.show()

# 1. Operasi SELECT -> memilih kolom tertentu
df.select("EmployeeName", "Salary").show()

# 2. Operasi FILTER -> menampilkan karyawan dengan gaji lebih dari 4000
df.filter(df.Salary > 4000).show()

# 3. Operasi GROUPBY + MEAN -> rata-rata gaji per departemen
df.groupBy("Department").agg(F.mean("Salary").alias("Average_Salary")).show()

# 4. Operasi GROUPBY + MAX -> gaji tertinggi per departemen
df.groupBy("Department").agg(F.max("Salary").alias("Max_Salary")).show()

# 5. Operasi GROUPBY + SUM -> total gaji per departemen
df.groupBy("Department").agg(F.sum("Salary").alias("Total_Salary")).show()
```

# 3. Bekerja dengan Tipe Data Kompleks

Tugas 3

1. ArrayType

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, ArrayType
from pyspark.sql import functions as F

# Tambahkan data dengan kolom skill (Array)
data_complex = [
    ('James', 'Sales', 3000, ['Negotiation', 'Communication']),
    ('Michael', 'Sales', 4600, ['Presentation']),
    ('Robert', 'Sales', 4100, ['Networking', 'Communication']),
    ('Maria', 'Finance', 3000, ['Accounting', 'Excel']),
    ('Siti', 'IT', 4500, ['Python', 'Spark', 'SQL'])
]

# Definisikan schema dengan ArrayType
schema = StructType([
    StructField("EmployeeName", StringType(), True),
    StructField("Department", StringType(), True),
    StructField("Salary", IntegerType(), True),
    StructField("Skills", ArrayType(StringType()), True)
])

# Membuat DataFrame
df_complex = spark.createDataFrame(data_complex, schema=schema)

# Menampilkan DataFrame
df_complex.show(truncate=False)

# Operasi explode -> "meledakkan" array jadi baris terpisah
df_complex.withColumn("Skill", F.explode("Skills")).select("EmployeeName", "Skill").show()
```

2. StructType (Nested Object)

```python
# Data dengan nested struct (alamat)
data_struct = [
    ('Ali', 'IT', 4000, ('Jakarta', 'DKI Jakarta')),
    ('Budi', 'HR', 3900, ('Bandung', 'Jawa Barat'))
]

schema_struct = StructType([
    StructField("EmployeeName", StringType(), True),
    StructField("Department", StringType(), True),
    StructField("Salary", IntegerType(), True),
    StructField("Address", StructType([
        StructField("City", StringType(), True),
        StructField("Province", StringType(), True)
    ]), True)
])

df_struct = spark.createDataFrame(data_struct, schema=schema_struct)
df_struct.show(truncate=False)

# Mengakses field dalam struct
df_struct.select("EmployeeName", "Address.City", "Address.Province").show()
```

3. MapType (Key-Value Pair)

```python
from pyspark.sql.types import MapType

data_map = [
    ('Citra', 'HR', 3500, {"Hobby": "Reading", "Status": "Full-time"}),
    ('Dina', 'Marketing', 4200, {"Hobby": "Traveling", "Status": "Contract"})
]

schema_map = StructType([
    StructField("EmployeeName", StringType(), True),
    StructField("Department", StringType(), True),
    StructField("Salary", IntegerType(), True),
    StructField("ExtraInfo", MapType(StringType(), StringType()), True)
])

df_map = spark.createDataFrame(data_map, schema=schema_map)
df_map.show(truncate=False)

# Mengakses nilai dalam map
df_map.select("EmployeeName", F.col("ExtraInfo.Hobby"), F.col("ExtraInfo.Status")).show()
```

# 4. Operasi Data Lanjutan

```python
from pyspark.sql import Window
from pyspark.sql import functions as F

# 1. WindowSpec berdasarkan Department dan urutkan Salary menurun
windowDept = Window.partitionBy("Department").orderBy(F.desc("Salary"))

# 2. Tambahkan ranking berdasarkan gaji dalam setiap departemen
df_rank = df.withColumn("Rank_in_Dept", F.rank().over(windowDept))
df_rank.show()

# 3. WindowSpec untuk running total gaji per departemen
windowDeptRows = Window.partitionBy("Department").orderBy("Salary").rowsBetween(Window.unboundedPreceding, 0)

df_running = df.withColumn("Running_Total_Salary", F.sum("Salary").over(windowDeptRows))
df_running.show()

# 4. Running total gaji keseluruhan (tanpa partition)
windowAll = Window.orderBy("Salary").rowsBetween(Window.unboundedPreceding, 0)
df_total = df.withColumn("Running_Total_All", F.sum("Salary").over(windowAll))
df_total.show()
```
